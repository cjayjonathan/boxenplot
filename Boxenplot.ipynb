```python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.boxenplot(x=tips["total_bill"])
```

```python
# Drawing a vertical boxen plot grouped by a categorical variable
ax = sns.boxenplot(x="day", y="total_bill", data=tips)
```

```python
# Drawing a letter value plot with nested grouping by two categorical variables
ax = sns.boxenplot(x="day", y="total_bill", hue="smoker",
                   data=tips, palette="Set3")
```

```python
# Drawing a boxen plot with nested grouping when some bins are empty
ax = sns.boxenplot(x="day", y="total_bill", hue="time",
                   data=tips, linewidth=2.5)
```

```python
# Control box order by passing an explicit order
ax = sns.boxenplot(x="time", y="tip", data=tips,
                   order=["Dinner", "Lunch"])
```

```python
# Draw a boxen plot for each numeric variable in a DataFrame using the IRIS dataset
iris = sns.load_dataset("iris")
ax = sns.boxenplot(data=iris, orient="h", palette="Set2")
```

```python
# Using stripplot() to show the datapoints on top of the boxes with showfliers=True
ax = sns.boxenplot(x="day", y="total_bill", data=tips,
                   showfliers=True)
ax = sns.stripplot(x="day", y="total_bill", data=tips,
                   size=4, color=".26")
```

```python
# Using stripplot() to show the datapoints on top of the boxes with showfliers=False
ax = sns.boxenplot(x="day", y="total_bill", data=tips,
                   showfliers=False)
ax = sns.stripplot(x="day", y="total_bill", data=tips,
                   size=4, color=".26")
```

```python
# Use catplot() to combine boxenplot() and a FacetGrid. This allows grouping within additional categorical variables. 
# Using catplot() is safer than using FacetGrid directly, 
# as it ensures synchronization of variable order across facets.
g = sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="boxen",
                height=4, aspect=.7);
```
